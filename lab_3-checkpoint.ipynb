```python
statistics = df.describe()
print(statistics)
```

```python
response_counts = df['Response'].value_counts()
response_counts.plot(kind='bar', color=['green', 'red'])
plt.title('Total Number of Responses')
plt.xlabel('Response')
plt.ylabel('Count')
plt.show()
```

```python
response_by_channel = df.groupby(['Sales Channel', 'Response']).size().unstack()
response_by_channel['Response Rate'] = response_by_channel['Yes'] / response_by_channel.sum(axis=1)
sns.barplot(x=response_by_channel.index, y='Response Rate', data=response_by_channel)
plt.title('Response Rate by Sales Channel')
plt.xlabel('Sales Channel')
plt.ylabel('Response Rate')
plt.show()
```

```python
response_by_total_claim = df.groupby('Total Claim Amount')['Response'].value_counts(normalize=True).unstack()
response_by_total_claim['Response Rate'] = response_by_total_claim['Yes']
plt.figure(figsize=(10, 6))
sns.scatterplot(x=response_by_total_claim.index, y='Response Rate', data=response_by_total_claim)
plt.title('Response Rate by Total Claim Amount')
plt.xlabel('Total Claim Amount')
plt.ylabel('Response Rate')
plt.show()
```

```python
response_by_income = df.groupby('Income')['Response'].value_counts(normalize=True).unstack()
response_by_income['Response Rate'] = response_by_income['Yes']
plt.figure(figsize=(10, 6))
sns.scatterplot(x=response_by_income.index, y='Response Rate', data=response_by_income)
plt.title('Response Rate by Income')
plt.xlabel('Income')
plt.ylabel('Response Rate')
plt.show()
```
